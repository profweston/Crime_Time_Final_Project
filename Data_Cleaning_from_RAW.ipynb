```python
# Add dependencies
import pandas as pd
import numpy as np
import datetime as dt

# Files to load
file_to_load = "Crimes_RAW.csv"

# Read the CSV into a DataFrame
raw_crime_df = pd.read_csv(file_to_load)
raw_crime_df.head()
```

```python
# Drop Case Number, Block, IUCR, Description, Beat, Community Area, FBI Code, X & Y Coordinate, Updated On and Location.
raw_crime_df.drop(raw_crime_df.columns[[1, 3, 4, 6, 10, 12, 13, 14, 15, 16, 18, 21]], axis=1, inplace=True)
raw_crime_df.head()
```

```python
# Remove NaN values
raw_crime_df.replace('NaN', np.NaN)
raw_crime_df = raw_crime_df.dropna()
raw_crime_df.head()
```

```python
# Split date column into Dates & Time
raw_crime_df["Dates"]=pd.to_datetime(raw_crime_df["Date"]).dt.date
raw_crime_df["Time"]=pd.to_datetime(raw_crime_df["Date"]).dt.time
raw_crime_df.head()
```

```python
# Remove the Date column
raw_crime_df.drop(raw_crime_df.columns[[1]], axis=1, inplace=True)
raw_crime_df.head()
```

```python
# Determine datatypes
raw_crime_df.dtypes
```

```python
# Drop dates from before 2015
raw_crime_df['Dates'] = pd.to_datetime(raw_crime_df['Dates'])
filtered_crime_df = raw_crime_df[~(raw_crime_df['Dates'] < '2015-01-01')]
filtered_crime_df.head()
```

```python
# export the clean dataset
filtered_crime_df.to_csv('clean_crimewk2.csv')
```
